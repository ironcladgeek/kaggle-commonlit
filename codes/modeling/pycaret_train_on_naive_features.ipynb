```python
%%capture
import pandas as pd
from pycaret.regression import *
```

```python
data = pd.read_csv('../../data/processed/train_naive_features.csv')
data.shape
```

```python
data.head()
```

```python
exp = setup(data=data, target='target', session_id=42, ignore_features=['id'], 
            fold=5, n_jobs=None, normalize=True,)
#             combine_rare_levels = True, rare_level_threshold = 0.05,
#             remove_multicollinearity = True, multicollinearity_threshold = 0.98, train_size=0.85)
```

```python
best = compare_models()
```

```python
gbr = create_model('gbr')
```

```python
tuned_gbr = tune_model(gbr, optimize='RMSE')
```

```python
predict_model(gbr);
```

```python
et = create_model('et')
```

```python
tuned_et = tune_model(et, optimize='RMSE')
```

```python
predict_model(et);
```

```python
lgbm = create_model('lightgbm')
```

```python
tuned_lgbm = tune_model(lgbm, optimize='RMSE')
```

```python
blended_model = blend_models(estimator_list=[gbr, et, tuned_lgbm], optimize='RMSE')
```

```python
predict_model(blended_model);
```

```python
final_model = finalize_model(blended_model)
```

```python
save_model(final_model, '../../models/blended_model_trained_on_naive_features')
```
